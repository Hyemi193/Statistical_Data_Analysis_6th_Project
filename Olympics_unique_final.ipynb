산점도 1 : 나라별 인구 수와 메달 수<hr>

```python
# 국가별 데이터 준비
datasets = [USA_only, China_only, Japan_only, Russia_only]
countries = ["USA", "China", "Japan", "Russia"]
colors = ["red", "darkorange", "indianred", "darkgreen"]

# 그래프 그리기 (산점도를 파란색 단색으로 설정 & USA 범위 조정)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()
fig.suptitle("나라별 인구 수와 메달 수", fontsize=16, fontweight="bold", y=1.00)

for i, (data, country) in enumerate(zip(datasets, countries)):
    x = data["Real_Medal"]
    y = data["Population"]
    
    # 상관계수 계산
    correlation = np.corrcoef(x, y)[0, 1]

    # 산점도 (파란색 단색)
    axes[i].scatter(x, y, c="blue", alpha=0.7, edgecolors='k')

    # 회귀선 추가
    sns.regplot(
        x=x, 
        y=y, 
        scatter=False,  # 산점도는 이미 그렸으므로 제외
        line_kws={"color": "red", "linestyle": "-"}, 
        ci=95,  # 95% 신뢰구간 포함
        ax=axes[i]
    )

    # USA의 경우 y축 범위 조정 (상위 95% 값까지만 표시)
    if country == "USA":
        max_y = np.percentile(y, 95)
        axes[i].set_ylim(min(y), max_y)

    # 그래프 제목 및 라벨 설정
    axes[i].set_xlabel("메달 수")
    axes[i].set_ylabel("인구 수")
    axes[i].set_title(f"{country}\n상관계수 값 : {correlation:.2f}", pad=10, fontweight="bold")

# 전체 레이아웃 조정
plt.tight_layout()
plt.show()
```

```python
# 그래프 그리기
plt.figure(figsize=(5, 4))

# 산점도 
scatter = plt.scatter(
    south_Korea_only['Real_Medal'], 
    south_Korea_only['Population'], 
    c="navy",
    alpha=0.7, 
    edgecolors='k'
)

# 회귀선 및 신뢰구간 추가 
sns.regplot(
    x=south_Korea_only['Real_Medal'], 
    y=south_Korea_only['Population'], 
    scatter=False,  # 기본 산점도는 비활성화 (위에서 따로 그리므로 중복 방지)
    line_kws={"color": "red", "linestyle": "-"}, 
    ci=95  # 95% 신뢰구간 포함
)

# 라벨 및 제목 추가
plt.xlabel("메달 수")
plt.ylabel("인구 수")
plt.title("South Korea\n상관계수 값 : 0.86", pad=15, fontweight='bold')

# 그래프 표시
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# QQ Plot 그리기 함수
def plot_qq(data, title):
    fig, ax = plt.subplots(figsize=(6, 6))
    stats.probplot(data, dist="norm", plot=ax)
    ax.get_lines()[1].set_color('red')  # 정규선 색상 변경
    ax.get_lines()[1].set_linewidth(2)  # 선 두께 조정
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xlabel("이론적 분위수", fontsize=12)
    ax.set_ylabel("실제 분위수", fontsize=12)
    plt.grid(alpha=0.3)
    plt.show()

# 참가 인원 QQ Plot
plot_qq(olympicsDF["Total_Athletes"], "참가 인원의 QQ Plot")

# 인구수 QQ Plot
plot_qq(olympicsDF["Population"], "인구 수의 QQ Plot")
```
